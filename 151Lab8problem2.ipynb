```python
y=symbols('y',real=True)
f=(y**2*(y+1)**3)/((y-2)**2*(y-4)**4) 
plot(f,(y,-10,10),ylim=[-10,10])
print('There appear to be no local extrema or inflection points.') 
df=diff(f,y)
print('The derivative of f is',df.simplify()) 
cvals=solve(df,y)
print(cvals)
print(cvals[2].simplify())
# cvalsdec=[float(i) for i in cvals]
# print('The critical values are',cvalsdec)
```
