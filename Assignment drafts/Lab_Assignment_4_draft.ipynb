3a

```python
## We just want f and g have one intersection
x = symbols('x')
a = symbols('a')
f = x**2
g = a*x**2 -2*x +2
df = diff(f,x)
dg = diff(g,x)
soln = solve([f-g,df-dg ], a,x)
a = soln[0][0]
print("a =", a)
```

3b

```python
g = a*x**2 -2*x +2
x_com = solve(f-g,x)[0]
df = diff(f,x)
slope = df.subs(x,x_com)
y_com = f.subs(x,x_com)
print("common pt is (",x_com,",",y_com,")")
print("Tangent line is y=",slope,"(x-",x_com,")+",y_com)
```
