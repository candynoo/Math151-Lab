```python
# Problem 1
# part(a)
x = symbols('x')
domain = reduce_inequalities(x-x**2>=0)
print(domain)

# part(b)
f = x*sqrt(x-x**2)
df = diff(f,x)
print(df.simplify())

# part(c)
plot(f,(x,0,1))

# part(d)
ext = solve(df,x)
M = f.subs(x,ext[0])
print("abs max at",ext[0], "is",M)
print("abs min at 0 and 1, value is 0")
```

```python
# Problem 2
# part(a)
x = symbols('x', real=true)
f = exp(x) + exp(-3*x)
df = diff(f,x)
critical = solve(df,x)
print(critical[0], critical[0].evalf())

# part(b)
df0 = df.subs(x,0)
df1 = df.subs(x,1)
print(df0.evalf(), df1.evalf())
m = f.subs(x,critical[0])
print("f decreasing then increasing, f reaches its abs min at", critical[0],"with value", m.evalf() )
print("f(0)=", f.subs(x,0))
print("f(1)=", f.subs(x,1).evalf())
print("f reaches its abs max at", f.subs(x,1).evalf())

# part(c)
plot(f,(x,0,1))



```

```python
# Problem 3
# part(a)
x = symbols('x')
c = 3
f = sin(x+sin(c*x))
plot(f,(x,0,2*pi))
print("There appeared to be 4 local max extrema and 4 local min extrema")

# part(b)
df = diff(f,x)
print(df)

# part(c)
plot(df,(x,0.5,0.7),ylim=(-0.0025,0.0025))

# part(d)
print("From the graph, there are 3 critical values in (0.5,0.7)")
c1 = nsolve(df, x, 0.6)
c2 = nsolve(df, x, 0.64)
c3 = nsolve(df, x, 0.68)
print(c1,c2,c3)

# part(e)
plot(df, (x,5.5,5.7),ylim=(-0.0025,0.0025))
print("From the graph, there are 3 critical values in (5.5,5.7)")
c4 = nsolve(df, x, 5.6)
c5 = nsolve(df, x, 5.65)
c6 = nsolve(df, x, 5.67)
print(c4,c5,c6)

# part(f)
print("Actually, there are 12 critical values (6 max and 6 min)")
```
