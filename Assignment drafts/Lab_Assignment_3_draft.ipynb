```python
d = symbols('d')
M = 0.1
g = 9.81
C = 1
rho = 1000
beta = 10 * pi / 180
theta = 10 * pi / 180 
V = sqrt(16 * M * g / (pi * C * rho * d**2))/sqrt(1-(8*M *(tan(beta))**2 / (pi * d**3 *C *rho *sin(theta)) ))
graph = plot_parametric(d,V,(d,0,0.1), ylim=[0.7,0.9])
a = V.subs(d,0.091).evalf()
b = V.subs(d,0.092).evalf()
print(a,b)
print("Intermediate Value Thm")
soln = solve(V-0.8,d)
print(soln)
```
