```python
t=symbols('t')
f = t**2
g = t**3 -3*t
r = Matrix([f,g])
plot_parametric(f,g,(t,-2,2))
print("intersect at (3,0)")
```

```python
t_soln = solve([f-3,g],t)
t_soln
```

```python
df = diff(f,t)
dg = diff(g,t)
dydx = dg/df
m1 = dydx.subs(t,t_soln[0][0])
m2 = dydx.subs(t,t_soln[1][0])
v1 = Matrix([1,m1])
v2 = Matrix([1,m2])
print(v1,v2)
```

```python
angle = acos(v1.dot(v2) / (v1.norm()*v2.norm()))
angle
```

```python
X = 3*t**2 +1
Y = 2*t**3+1
dX = diff(X,t)
dY = diff(Y,t)
dydx = dY/dX
t0 = solve(dydx * (X-4) - Y +3 , t)
print(t0)
m1 = t0[0]
m2 = t0[1]
x1 = X.subs(t,t0[0])
x2 = X.subs(t,t0[1])
y1 = Y.subs(t,t0[0])
y2 = Y.subs(t,t0[1])
print("points are (",x1,",",y1,") with slope ", m1, "and (",x2,",",y2,") with slope ", m2)
```

```python
x,y = symbols('x,y')
tan1 = m1*(x-x1) + y1
tan2 = m2*(x-x2) + y2
graph1 = plot_parametric(X,Y,(t,-2.5,2.5),xlim=[0,15],ylim=[-15,10],show = False)
graph2 = plot((tan1,(x,0,15)),(tan2,(x,0,15)),show = False)
graph1.extend(graph2)
graph1.show()
```
