```python
#Problem 1
# part(a)
x = symbols('x', positive = True)
f = 2* asin(x)
g = acos(1-2*x**2)
h = f-g
print(diff(h,x).simplify())

# part(b)
print("f-g is a constant")

# part(c)
print("f(0)-g(0) is ",h.subs(x,0), ", so f-g=0 for all positive x, f=g")
```

```python
#Problem 4
# part(a)
x = symbols('x', real = True)
f = x*sin(x-1)
g = 2*x**2-x-1

# part(b)
f1 = limit(f,x,1)
g1 = limit(g,x,1)
print(f1, g1)

# part(c)
df = diff(f,x)
dg = diff(g,x)
tan_f = df.subs(x,1)
tan_g = dg.subs(x,1)
tan1 = tan_f *(x-1)
tan2 = tan_g *(x-1)
plot((f,(x,0.5,1.5)),(g,(x,0.5,1.5)),(tan1,(x,0.5,1.5)),(tan2,(x,0.5,1.5)))
print("L_f = ", tan1)
print("L_g = ", tan2)

# part(d)
print(limit(f/g,x,1))
print("The limit = f'(1)/g'(1)")


```
